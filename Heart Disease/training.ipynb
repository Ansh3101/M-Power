```python
df = pd.read_csv('Data.csv')

X = df

Y = X.pop('HeartDisease')
```

```python
pp = hp.preProcessing(df=X)
```

```python
Y = pd.Categorical(Y).codes
```

```python
set(df["PhysicalHealth"])
```

```python
X['Smoking'] = pd.Categorical(X['Smoking']).codes
X['AlcoholDrinking'] = pd.Categorical(X['AlcoholDrinking']).codes
X['Stroke'] = pd.Categorical(X['Stroke']).codes
X['DiffWalking'] = pd.Categorical(X['DiffWalking']).codes
X['Sex'] = pd.Categorical(X['Sex']).codes
X['AgeCategory'] = pd.Categorical(X['AgeCategory']).codes
X['Race'] = pd.Categorical(X['Race']).codes
X['Diabetic'] = pd.Categorical(X['Diabetic']).codes
X['PhysicalActivity'] = pd.Categorical(X['PhysicalActivity']).codes
X['GenHealth'] = pd.Categorical(X['GenHealth']).codes
X['Asthma'] = pd.Categorical(X['Asthma']).codes
X['KidneyDisease'] = pd.Categorical(X['KidneyDisease']).codes
X['SkinCancer'] = pd.Categorical(X['SkinCancer']).codes

X
```

```python
num_cols = pp.dtype_select(np.number)
num_cols
```

```python
X = pp.minMax(num_cols)
X
```

```python
X.describe()
```

```python
import catboost as cb
from sklearn.model_selection import train_test_split
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.2)
```

```python
model = cb.CatBoostClassifier(iterations=5000, learning_rate=1e-2)
model.fit(train_X, train_Y)
```

```python
pred = model.predict(test_X)
```

```python
list(pred), list(test_Y)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(pred, test_Y)
```
